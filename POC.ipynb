```python
import json
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Optional
from fastapi import FastAPI
sns.set_style('white')
%matplotlib inline
```

```python
with open("products.json",encoding='utf-8') as file:
    dataProd = json.load(file)

with open("order.json",encoding='utf-8') as file:
    dataOrders = json.load(file)

with open("crop.json",encoding='utf-8') as file:
    dataCrop = json.load(file)
```

```python
crops = dict()
for crop in dataCrop['data']:
    cropName = crop['crop_name']
    if cropName not in crops:
        crops[crop['crop_name']]=0
# crops
```

```python
data = list()
crop_matrix = list()
for item in dataProd['data']:
    crops_copy = crops.copy()
    dataRow =     {
            'fasalSetuProductId':item['fasalSetuProductId'],
            'productCategory':item['productCategory'][0]['id'],
            'productSubCategories':item['productSubCategories'][0]['id'],
            'productShortName':item['productShortName']
        };
    for i in item['productCropName']:
        crops_copy[i]=1
    crop_matrix.append(list(crops_copy.values()))
    dataRow.update(crops_copy)
    data.append(dataRow)
df = pd.DataFrame(data) 
df.to_csv('product.csv')
crop_matrix = np.array(crop_matrix )
# crop_matrix
# df
```

```python
indicies = pd.Series(df.index,index = df['productShortName']).drop_duplicates()
```

```python
indicies
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
sig = sigmoid_kernel(crop_matrix,crop_matrix)
```

```python
sig[0]
```

```python
def give_recomendation(prodId,sig=sig):
    idx = indicies[prodId]
    sig_scores = list(enumerate(sig[idx]))
    sig_scores = sorted(sig_scores,key = lambda x : x[1],reverse = True)
    sig_scores = sig_scores
    prod_indexes = [i[0] for i in sig_scores][:10]
    return df['productShortName'].iloc[prod_indexes].tolist()
```

```python
give_recomendation('APPU 250 GM - TEST')
```

```python
import pickle
pickle_out = open("sig.pkl","wb")
pickle.dump(sig, pickle_out)
pickle_out.close()
```
